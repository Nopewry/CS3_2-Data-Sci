```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Customer.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
# first step
att = data[["age", "salary"]]
allclass = data["purchased"]

att_train, att_test, class_train, class_test = train_test_split(att, allclass, random_state = 0, train_size = .7)


#secound step
scaler = StandardScaler()
scaler.fit(att_train)
att_train[[ "age", "salary"]] = scaler.transform(att_train)

model = KNeighborsClassifier(n_neighbors = 3)
model.fit(att_train, class_train)

# third step
model.score(scaler.transform(att_test), class_test)
```

```python
scaler.inverse_transform(att_train)
# att_train[[ "age", "salary"]]
```

```python
model.predict(att_train)
```

```python
result = pd.concat([att_train, class_train], axis = 1)
result["predict"] = model.predict(att_train)
result
```
