```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('vgsales.csv',encoding = 'iso-8859-1')
```

```python
sns.scatterplot(data = data, x = "JP_Sales", y = "Other_Sales")
```

```python
df2 = data[["JP_Sales", "Other_Sales"]].dropna()
df2.head()
```

```python
model = KMeans(n_clusters = 4, random_state = 0)
model.fit(df2)
```

```python
plt.figure(figsize = [10, 8])
sns.scatterplot(data = df2, x = "JP_Sales", y = "Other_Sales", hue = model.labels_, palette = "Set1")
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], color = "k", marker = "*")
```

```python
model.predict([[0, 1],[2, 4]])
```
